```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing data
autos=pd.read_csv("Automobile_data.csv")
```

```python
#viewing data
pd.set_option("display.max_rows",None)
autos
```

```python
#Automobile ratings according to wheel base in descending order, it also shows the correlation between wheel-base and length
autos.sort_values(by="wheel-base", ascending=False,ignore_index=True)
```

```python
#Automobile ratings according to horsepower in descending order
autos.sort_values(by="horsepower", ascending=False,ignore_index=True)
```

```python
#Visualizing according to price
plt.figure(figsize=(12,8))
plt.barh(autos["company"],autos["price"],color="blue")
plt.title("Average Price Rating")
plt.xlabel("Price")
plt.ylabel("Company")
```

```python
#Visualizing according to price
plt.figure(figsize=(12,8))
plt.barh(autos["company"],autos["horsepower"],color="red")
plt.title("Average Horsepower Rating")
plt.xlabel("Horsepower")
plt.ylabel("Company")
```

```python
autos_correlation=autos.corr()
autos_correlation
```

```python
#Heatmap for correlation
plt.figure(figsize=(15,10))
sns.heatmap(autos_correlation, annot=True)
plt.title("Correlation for Numeric Features in Dataset")
plt.xlabel("Automobile Features")
plt.ylabel("Automobile Features")
```

```python
#it shows there is high correlation between Automobile Horsepower and Automobile Price, 
#so we can conclude that the higher the horsepower of the car the more expensive it becomes 
plt.figure(figsize=(15,10))
sns.regplot(x=autos["horsepower"],y=autos["price"], scatter_kws ={"color": "magenta"}, line_kws={"color":"blue"})
plt.title("Horsepower vs Price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
```
