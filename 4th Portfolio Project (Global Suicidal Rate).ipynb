```python
#Add new column named suicide percentage to get the actual percentage of people commiting suicide
df["suicide_percentage"]=df["suicides_no"]/df["population"]*100
```

```python
df
```

```python
#calling the country  and suicide percentage from 1987 to 2016
df[["country","suicide_percentage"]].groupby("country").sum().reset_index()
```

```python
#calling the country  and suicide percentage then sorting by suicide percentage in descending order
country_suicide_rating=df[["country","suicide_percentage"]].groupby("country").sum().reset_index().sort_values(by="suicide_percentage",ascending=False,ignore_index=True)
```

```python
#Top 10 countries with the highest suicide rate
country_suicide_rating.head(10)
```

```python
#least 10 countries with the lowest suicide percentage
country_suicide_rating.tail(10)
```

```python
yearly_suicide_rate=df[["year","suicide_percentage"]].groupby("year").sum().reset_index().sort_values(by="suicide_percentage",ascending=False,ignore_index=True)
```

```python
#it shows suicide rate was very high in 1995
yearly_suicide_rate.head(10)
```

```python
#it shows suicide rate was very low in 2016
yearly_suicide_rate.tail(10)
```

```python
plt.figure(figsize=(12,15))
plt.barh(yearly_suicide_rate["year"],yearly_suicide_rate["suicide_percentage"],color="magenta")
plt.title("Suicide Percentage/ Year")
plt.xlabel("Year")
plt.ylabel("Suicide Percentage")
```

```python
#it shows the male gender are commiting suicide more
gender_rate=df.groupby("sex").sum().reset_index()
gender_rate
```

```python
Gender_Suicide_Rate=[Female,Male]
Gender_Suicide_Rate
```

```python
plt.figure(figsize=(20,10))
plt.pie(Gender_Suicide_Rate,
        labels=["Female","Male"],
        autopct=("%1.1f%%")
        )
plt.legend(loc=("lower right"))
plt.title("Suicide Rate Based On Gender")
```
